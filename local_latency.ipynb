```python
import time
import queue
import asyncio
import matplotlib.pyplot as plt
```

```python
n = 50000
```

```python
# Test latency of list append/pop
lst = []
def put():
    lst.append(1)
    return time.time()
def get():
    lst.pop(-1)
    return time.time()
start = time.time()
data = [-(put() - get()) for i in range(n)]
end = time.time()
print('time elapsed:', end - start)
plt.plot(data)
```

```python
# Test latency of asyncio.Queue put/get

q = asyncio.Queue()
async def put():
    await q.put(1)
    return time.time()
async def get():
    await q.get()
    return time.time()
start = time.time()
data = [-(await put() - await get()) for i in range(n)]
end = time.time()
print('time elapsed:', end - start)
plt.plot(data)
```

```python
# Test latency of queue.Queue put/get
q = queue.Queue()
q.put(1)
def put():
    q.put(1)
    return time.time()
def get():
    q.get()
    return time.time()
start = time.time()
data = [-(put() - get()) for i in range(n)]
end = time.time()
print('time elapsed:', end - start)
plt.plot(data)
```
